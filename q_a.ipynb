```python
# solve e^x = x
# using Newton's method
# f(x) = e^x - x

import numpy as np
import matplotlib.pyplot as plt

def newton(x0, f, df, tol, maxiter):
    x = x0
    interset = []
    for i in range(maxiter):
        x = x - f(x)/(df(x)+1e-10)
        if abs(f(x)) < tol:
            interset.append(x)
    return interset
```

```python
x0 = 0.1
f = lambda x: np.sin(x) - np.log(x)
df = lambda x: np.cos(x) - 1/(x+1e-10)
tol = 1e-12
maxiter = 100
x = newton(x0=x0, f=f, df=df, tol=tol, maxiter=maxiter)
print(len(x))
```

```python
fig, ax = plt.subplots()
x = np.linspace(0.1, 5, 10_000)
ax.plot(x, f(x), label='$f(x)=\sin(x)-ln(x)$')
ax.plot(x, np.sin(x), label='$\sin(x)$')
ax.plot(x, np.log(x), label='$\ln(x)$')
# ax.plot(x, np.zeros_like(x))
ax.legend()
# ax.plot(x, df(x))
```
